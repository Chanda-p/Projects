#Hyperparameter tuning refers to the process of selecting the optimal values for the hyperparameters of a machine learning model.


#max depth -

max_depth is a parameter in decision tree algorithms that specifies the maximum depth or maximum number of levels allowed in the decision tree.

#min_samples_split - default=2

->min_samples_split is a parameter in decision tree algorithms
that determines the minimum number of samples required to split an internal node.
->It specifies the minimum number of samples that should be present in a node for the algorithm to consider splitting it further.

#min_samples_leaf - default = 1

-> min_samples_leaf is another parameter in decision tree algorithms that determines the minimum number of samples required to be present in a leaf node.
-> It specifies the minimum number of samples that should be in a leaf for the algorithm to consider it as a valid leaf node.



```python
#Hyperparameter Tuning
from sklearn.model_selection import GridSearchCV

# Define the parameter grid to search

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}
```

```python
# Create the Random Forest classifier
rf_classifier = RandomForestClassifier(random_state=42)
```

```python
# Perform grid search using cross-validation
grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid, cv=2)
grid_search.fit(X_train, y_train)
```

```python
# Print the best hyperparameters found
print("Best Hyperparameters: ", grid_search.best_params_)
best_dt = grid_search.best_estimator_
```

```python
# Builiding the random forest model again with best parameters

classifier1 = RandomForestClassifier(n_estimators = 200, min_samples_split = 10, min_samples_leaf = 2, max_depth = None)
classifier1.fit(X_train,y_train)
```

```python
y_pred = classifier1.predict(X_test)
```

```python
acc = accuracy_score(y_pred,y_test)
acc
```

```python
y_pred_train = classifier1.predict(X_train)
```

```python
acc2 = accuracy_score(y_pred_train,y_train)
acc2
```

We built a Random Forest classifier using the RandomForestClassifier from scikit-learn. The initial model achieved an accuracy of approximately 79.03% on the test set and 99.79% on the training set. However, these results might indicate overfitting, as the model performed significantly better on the training data than the test data.

To address this, we performed hyperparameter tuning using GridSearchCV to search for the best combination of hyperparameters for the Random Forest model. After the hyperparameter tuning, the best hyperparameters were found to be: 'n_estimators': 200, 'max_depth': None, 'min_samples_split': 10, and 'min_samples_leaf': 2. We then built a new Random Forest classifier using these best hyperparameters.

The tuned Random Forest model achieved an improved accuracy of approximately 80.10% on the test set, while the accuracy on the training set decreased to around 89.87%. This suggests that the model generalizes better to unseen data after hyperparameter tuning.

In conclusion, by carefully selecting the best hyperparameters through GridSearchCV, we were able to improve the model's performance and reduce overfitting. The final Random Forest classifier with tuned hyperparameters achieved an accuracy of approximately 80.10% on the test data, making it a more robust model for predicting customer churn. However, further analysis, feature engineering, and evaluation of other metrics may be necessary for a comprehensive understanding of the model's performance in a real-world application.
