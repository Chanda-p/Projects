# Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# Split the dataset into features (X) and target variable (y)
X = data.drop('target', axis=1)
y = data['target']
```

```python
# Divide the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create and fit the logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Predict values on the test set
y_pred = model.predict(X_test)

# Calculate the confusion matrix
confusion_mat = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(confusion_mat)

# Calculate the accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy Score:", accuracy)
```

The additional message  received is a warning message indicating that the logistic regression model did not converge to a solution within the specified number of iterations.

Adjusting the max_iter parameter in the LogisticRegression() function to a higher value, such as max_iter=1000, will allow the algorithm to iterate more times and potentially converge to a solution. 

```python
# Create and fit the logistic regression model with max_iter=1000
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Predict values on the test set
y_pred = model.predict(X_test)

# Calculate the confusion matrix
confusion_mat = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(confusion_mat)

# Calculate the accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy Score:", accuracy)
```

The confusion matrix indicates the following:

True Positives (TP): 32

True Negatives (TN): 42

False Positives (FP): 9

False Negatives (FN): 8

Accuracy Score: 0.8131868131868132

The accuracy score is 0.8132, which means the model correctly classified approximately 81.32% of the instances in the test set.

```python
from sklearn.metrics import classification_report
```

```python
# Instantiate and train the Logistic Regression model
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
logreg_pred = logreg.predict(X_test)

# Instantiate and train the Decision Tree model
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)

# Instantiate and train the Random Forest model
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
```

```python
# Print confusion matrix for all classifiers
classifiers = [logreg, dt, rf]
model_names = ['Logistic Regression', 'Decision Tree', 'Random Forest']
for model, name in zip(classifiers, model_names):
    pred = model.predict(X_test)
    cm = confusion_matrix(y_test, pred)
    print(f"\nConfusion Matrix - {name}:")
    print(cm)
```

```python
# Print classification report for all classifiers
for model, name in zip(classifiers, model_names):
    pred = model.predict(X_test)
    report = classification_report(y_test, pred)
    print(f"\nClassification Report - {name}:")
    print(report)
```

```python
# Calculate recall, precision, and F1 score for all models
for model, name in zip(classifiers, model_names):
    pred = model.predict(X_test)
    report = classification_report(y_test, pred, output_dict=True)
    recall = report['1']['recall']
    precision = report['1']['precision']
    f1_score = report['1']['f1-score']
    print(f"\nMetrics - {name}:")
    print(f"Recall: {recall}")
    print(f"Precision: {precision}")
    print(f"F1 Score: {f1_score}")
```

```python
# Visualize confusion matrix using heatmaps
plt.figure(figsize=(12, 4))
for i, model in enumerate(classifiers):
    pred = model.predict(X_test)
    cm = confusion_matrix(y_test, pred)
    ax = plt.subplot(1, len(classifiers), i+1)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')
    ax.set_title(model_names[i])
plt.tight_layout()
plt.show()
```

```python
# Select the best model based on accuracies
accuracies = [accuracy_score(y_test, model.predict(X_test)) for model in classifiers]
best_model_index = accuracies.index(max(accuracies))
best_model = classifiers[best_model_index]
best_model_name = model_names[best_model_index]
print(f"\nBest Model based on Accuracies: {best_model_name}")
```
